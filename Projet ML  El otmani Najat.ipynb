```python
from csv import reader
# Load a CSV file
def load_csv(filename):
  my_data = list()
  with open(filename, 'r') as file:
    csv_reader = reader(file)
    next(csv_reader)
    for row in csv_reader:
      if not row:
        continue
      my_data.append(row)
  return my_data
# Convert string column to float
def str_column_to_float(my_data, column):
    for row in my_data:
        if row[column] != '':
            row[column] = float(row[column].strip())
```

# Preprocessing
### **Scaling**

since my data has many different units we need to scale it in  a specific Range ,i choose to work with standardization helps ensure that the model considers all features fairly, reducing the risk of biased predictions due to feature scaling differences, and often improves model performance and  For instance, total_rooms may have values in the tens or hundreds of thousands, while median_income ranges from 0 to several thousand.
Standardization ensures that all features have a mean of 0 and a standard deviation of 1, making it easier for machine learning models to learn and compare the importance of each feature without being biased by differing scales.

**1. Standardize Data**

```python
def column_means(my_data):
    means = [0.0] * len(my_data[0])  # Initialize means for each column
    for i in range(len(my_data[0])):
        col_values = [float(row[i]) for row in my_data if isinstance(row[i], (int, float))]
        means[i] = sum(col_values) / float(len(col_values))  # Calculate mean for numeric values only
    return means
```

```python
means= column_means(my_data)
means
```

```python
from math import sqrt

def column_stdevs(my_data, means):
    stdevs = [0 for _ in range(len(my_data[0]))]  # Initialize standard deviations
    for i in range(len(my_data[0])):
        # Filter numeric values and calculate variance
        variance = [pow(row[i] - means[i], 2) for row in my_data if isinstance(row[i], (int, float))]
        stdevs[i] = sum(variance)
    stdevs = [sqrt(x / (float(len(my_data) - 1))) for x in stdevs]  # Calculate standard deviation
    return stdevs

# Example usage
stdevs = column_stdevs(my_data, means)
stdevs
# calculons la variance
i=1
variance = [pow(row[i]-means[i], 2) for row in my_data]
variance
```

```python
def calculate_variance(group):
    if len(group) == 0:
        return 0
    values = [row[-1] for row in group]
    mean_value = np.mean(values)
    variance = sum((x - mean_value) ** 2 for x in values) / len(values)
    return variance
def evaluate_split(left, right):
    total_size = len(left) + len(right)
    weighted_variance = (calculate_variance(left)*len(left) / total_size  +
                          calculate_variance(right)*len(right) / total_size )
    return weighted_variance
```

```python
def create_terminal_node(group):
    values = [row[-1] for row in group]
    return np.mean(values)
```

```python
def build_tree_regr(node, max_depth, min_size, depth):
    left, right = node['groups']
    del node['groups']
    
    # Check if we should create terminal nodes
    if not left or not right:
        node['left'] = node['right'] = create_terminal_node(left + right)
        return
    if depth >= max_depth:
        node['left'], node['right'] = create_terminal_node(left), create_terminal_node(right)
        return
    if len(left) <= min_size:
        node['left'] = create_terminal_node(left)
    else:
        node['left'] = get_best_split(left)
        build_tree_regr(node['left'], max_depth, min_size, depth + 1)
    if len(right) <= min_size:
        node['right'] = create_terminal_node(right)
    else:
        node['right'] = get_best_split(right)
        build_tree_regr(node['right'], max_depth, min_size, depth + 1)
```

```python
def train_regression_tree(my_data, max_depth, min_size):
    root = get_best_split(my_data)
    build_tree_regr(root, max_depth, min_size, 1)
    return root
```

The R2 (or R Squared) metric provides an indication of the goodness of fit of a set of predictions
to the actual values. In statistical literature this measure is called the coeficient of determination.
R² values range from 0 to 1
